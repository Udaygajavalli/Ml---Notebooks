```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
columns_labels = ['top',
 'platform_look',
 'ui/ux_improvements',
 'prefer_learnbydoing',
 'why_yes/no?',
 'widhya_effectiveness',
 'widhya_effectiveness?',
 'learnbydoing_interest',
 'is_widhya_helpful',
 'mission_completion_time',
 'mission_name',
 'mission_effectiveness',
 'mission_effectiveness?',
 'mission_difficulty',
 'mission_improvements',
 'mission_ratings',
 'microtasks_progress',
 'microstask_improvements',
 'overall_microtask_effectiveness_rate',
 'widhya_platform_satisfaction',
 'platform_recommendation',
 'learnt_something_new',
 'learnt_something_new?',
 'newtopics_recommendation',
'new_skills_learnt_after_widhyasmissions']
```

```python
df = pd.read_csv('Widhya.csv')
df.columns = columns_labels
df
```

```python
plt.figure(figsize=(10,5))
sns.set(style="dark")
sns.countplot(df.top,palette='twilight')
plt.xticks(rotation=45)
plt.xlabel('Top % in class')
plt.ylabel('Count of the people registered')
```

```python
plt.figure(figsize=(10,5))
sns.countplot(df.top,hue=df.mission_name,palette='rainbow')
plt.legend(loc='upper right')
plt.xticks(rotation=45)
plt.xlabel('Top % in class with Intern role registered')
plt.ylabel('Count of the people registered')
```
