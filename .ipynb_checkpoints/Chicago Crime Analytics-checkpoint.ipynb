```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Chicago_Crime_Detective.csv', index_col = 0)
```

```python
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
dd = df[pd.notnull(df['Date'])]

import math

median = math.floor(df['Date'].astype('int64').median())

result = np.datetime64(median, "ns") #unit: nanosecond
result
```

```python
df["month"] = df['Date'].map(lambda x: x.month)
df['month'].value_counts()
```

```python
df["day"] = df['Date'].map(lambda x: x.day_name())
df["day"].value_counts()
```

```python
df_a = df[df["Arrest"] == True]
df_a
```

```python
df_a["month"] = df_a['Date'].map(lambda x: x.month)
df_a['month'].value_counts().sort_index()
```

```python
df_a["year"] = df_a['Date'].map(lambda x: x.year)
df_a['year'].value_counts().sort_index()
```
